```python
import os
import librosa
import soundfile as sf

input_folder = 'sound_mnist/data/01'
output_folder = 'sound_mnist_16khz'

# Show the first 5 files in the input folder
print("Input files:")
for filename in os.listdir(input_folder)[:5]:
    print(f"  {filename}")
# Create the output folder if it doesn't exist 
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.wav'):
        # Load the audio file
        input_path = os.path.join(input_folder, filename)
        y, sr = librosa.load(input_path, sr=None)
        print(f"Processing {filename}...")
        # Resample the audio to 16kHz
        y_16k = librosa.resample(y, orig_sr=sr, target_sr=16000)
        
        # Save the resampled audio to the output folder
        output_path = os.path.join(output_folder, filename)
        sf.write(output_path, y_16k, 16000)

print("Conversion complete.")
```
