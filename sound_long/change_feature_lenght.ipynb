```python
import pandas as pd
import numpy as np

number_list = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]

# Load the CSV file and repeat each row 10 times for each number from 1 to 9
for i in number_list:
    df = pd.read_csv(f"/home/hastabbe/e-prop_master_nmbu/sound_fet/{i}_encoded_features.csv")
    df_expanded = df.loc[np.repeat(df.index, 10)].reset_index(drop=True).T
    df_expanded.to_csv(f"/home/hastabbe/e-prop_master_nmbu/sound_long/expanded_{str(i)}_feature.csv")
```

```python
df = pd.read_csv("/home/hastabbe/e-prop_master_nmbu/sound_fet/zeros.csv")
df_expanded = df.loc[np.repeat(df.index, 10)].reset_index(drop=True).T
df_expanded.to_csv("/home/hastabbe/e-prop_master_nmbu/sound_long/expanded_zero_feature.csv")
```
